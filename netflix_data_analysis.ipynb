```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
print("Dataset loaded!")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
type_count = df['type'].value_counts()
print(type_count)
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
top_genres = genres.value_counts().head(5)

print("Top Genres:")
print(top_genres)
```

```python
countries = df['country'].str.split(', ', expand=True).stack()
top_countries = countries.value_counts().head(5)

print("Top Countries:")
print(top_countries)
```

```python
import matplotlib.pyplot as plt

type_count.plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
print("=== KEY FINDINGS ===")
print(f"Total Content: {df.shape[0]}")
print(f"Movies: {type_count['Movie']}")
print(f"TV Shows: {type_count['TV Show']}")
print(f"Top Genre: {top_genres.index[0]}")
print(f"Top Country: {top_countries.index[0]}")
```
